```python
#age, sibsp, parch, fareí•­ëª©ì„ notnull()í•¨ìˆ˜ë¥¼ ì´ìš©í•´ì„œ nullì´ ì•„ë‹Œê²ƒë§Œ ì¶”ì¶œí•˜ì‹œì˜¤.
raw_data = raw_data[raw_data['age'].notnull()]
raw_data = raw_data[raw_data['sibsp'].notnull()]
raw_data = raw_data[raw_data['parch'].notnull()]
raw_data = raw_data[raw_data['fare'].notnull()]
raw_data.info()
```

```python
X_train, X_test, y_train, y_test = train_test_split(train_pre, #ë°ì´í„°í”„ë ˆìž„ì˜ ì†ì„±ë“¤
                 raw_data[['survived']], #ë§žì¶”ì–´ì•¼í•  ì •ë‹µ
                 test_size = 0.1, #í…ŒìŠ¤íŠ¸ë°ì´í„° 10%ì§€ì •, í›ˆë ¨ë°ì´í„° 90%í• ë‹¹
                 random_state = 13 #seedê°’
                )
```

```python
X_test = X_test.reset_index()
X_test = X_test.drop(['index'], axis=1)

y_train = y_train.reset_index()
y_train = y_train.drop(['index'], axis=1)

y_test = y_test.reset_index()
y_test = y_test.drop(['index'], axis=1)
```

```python
from sklearn.tree import DecisionTreeClassifier

#í•™ìŠµê¸° ê°ì²´ ìƒì„±
# íŒŒë¼ë©”í„°ë¥¼ ë¶„ì„ê°€ê°€ ìž„ì˜ë¡œ ë³€ê²½í•  ìˆ˜ ìžˆìŒ.
# íŒŒë¼ë©”í„°ë¥¼ ë¶„ì„ê°€ê°€ ë³€ê²½ê°€ëŠ¥, ë³€ê²½ë¶ˆê°€ëŠ¥
# ë¶„ì„ê°€ê°€ ë³€ê²½ê°€ëŠ¥í•œ íŒŒë¼ë©”í„°ë¥¼ "í•˜ì´í¼íŒŒë¼ë©”í„°"
tree_clf = DecisionTreeClassifier(max_depth = 6, random_state = 42)
#ê³µë¶€ì‹œí‚¤ê¸°
tree_clf.fit(X_train, y_train) 
#trainë°ì´í„°ë¥¼ ê°€ì§€ê³  ê³µë¶€í•œ ì ìˆ˜
tree_clf.score(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score #ì±„ì í•´ì£¼ëŠ” ë¼ì´ë¸ŒëŸ¬ë¦¬

y_pred = tree_clf.predict(X_test) #ë¬¸ì œë¥¼ ì£¼ê³  í’€ì–´ë³´ë¼ê³  í•¨. => ì •ë‹µ return!

#í…ŒìŠ¤íŠ¸ë°ì´í„°ë¥¼ í‘¼ ë‹µì•ˆê³¼ í•´ë‹µì„ ë¹„êµí•´ì„œ ì •í™•ë„ë¥¼ êµ¬í•´ë³´ìž.
accuracy_score(y_test, y_pred)
```
