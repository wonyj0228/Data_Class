<h2><코로나 바이러스란?></h2>

* 2019년 12월 중국 후베이성 우한시에서 발생하여 세계적으로 확산되어 새로 발견된 신종 코로나바이러스에 의한 호흡기 감염질환.

* 주된 전파경로는 감염자의 호흡기 침방울(비말)에 의한 전파.

* 사람 간에 전파되며, 대부분의 감염은 감염자가 기침, 재채기, 말하기, 노래 등을할 때 발생한 호흡기 침방울(비말)을다른사람이 밀접접촉(주로 2m 이내)하여 발생.

<h2><서울시 코로나19 확진자 현황></h2>

```python
raw_data = pd.read_csv('../data12/서울시 코로나19 확진자 현황.csv', engine = 'python')
# 사용할 데이터만 추출
raw_data = raw_data.drop(['환자번호','국적','환자정보','조치사항','상태','이동경로'], axis=1)
raw_data
```

```python
raw_data['지역'].unique()
#띄어쓰기 유무로 2개씩 생긴 지역이 있음
```

```python
raw_data.groupby('지역').count()
#띄어쓰기 있는 지역 확인
```

```python
replace_place = ['강남구 ','강북구 ','광진구 ','서초구 ','성북구 ','은평구 ','중구 ','타시도 ']
for x in replace_place:
    raw_data['지역'] = raw_data['지역'].replace(x,x.strip()) # 공백을 제거한 string 으로 재입력

raw_data.groupby('지역').count()
```

```python
null_place = ['타시도','기타']
for y in null_place:
    raw_data['지역'] = raw_data['지역'].replace(y,np.nan)

raw_data = raw_data[raw_data['지역'].notnull()]     
raw_data['지역'].unique()
```

```python
raw_data.groupby('지역').count()
```

<h2>서울시 지역구별 확진자 수</h2>

* 서울시에서는 송파구, 강서구, 강남구 순으로 지역감염이 많이 발생하고 있다.
* 지역감염이 가장 적은 지역구는 금천구, 중구, 종로구 순으로 정리할 수 있다.


```python
f, ax = plt.subplots(figsize = (18,10))
sns.countplot('지역', data=raw_data, ax=ax)
```

```python
songpa = raw_data[raw_data['지역'] == '송파구'].groupby('접촉력').count()
gangseo = raw_data[raw_data['지역'] == '강서구'].groupby('접촉력').count()
gangnam = raw_data[raw_data['지역'] == '강남구'].groupby('접촉력').count()
songpa
```

```python
songpa_graph = songpa.sort_values(by=['연번'],ascending = False).head(10)
gangseo_graph = gangseo.sort_values(by=['연번'],ascending = False).head(10)
gangnam_graph = gangnam.sort_values(by=['연번'],ascending = False).head(10)
```

```python
plt.figure(figsize=(10,10))
plt.subplot(111)
wedgeprops={'width': 0.7, 'edgecolor': 'white', 'linewidth': 3}
label = ['기타 확진자 접촉','감염경로 조사중','타시도 확진자 접촉','해외유입','강남구 럭키사우나 관련','강남구 어학원 관련Ⅰ','강남구 소재 직장3 관련','강남서초 지인모임','성북구 사랑제일교회 관련','강남구 소재 사우나(타시도 일가족 관련)']

pie = plt.pie(songpa_graph,
        autopct='%1.2f%%',
        wedgeprops=wedgeprops
       )
plt.legend(pie,labels)
```
