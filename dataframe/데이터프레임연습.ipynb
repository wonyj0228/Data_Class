```python
import pandas as pd
```

```python
df_2010 = pd.read_csv("../data/2010.csv", encoding='utf-8')
df_2011 = pd.read_csv("../data/2011.csv", encoding='utf-8')
df_2012 = pd.read_csv("../data/2012.csv", encoding='utf-8')
```

```python
df_2010
```

```python
df_2011
```

```python
df_2012
```

```python
total_jumsu = pd.concat([df_2010, df_2011, df_2012], ignore_index=True)  #합치기. 합칠때 기존 인덱스 무시하기.
```

```python
total_jumsu
```

```python
total_jumsu['국어']
```

```python
total_jumsu.iloc[0]  #첫번째 인스턴스의 행을 가져옴. 특정인덱스로 슬라이싱을 함. loc=location
```

```python
total_jumsu.iloc[0:2] #0~1까지 범위로 슬라이싱
```

```python
total_jumsu.iloc[:,0]  #iloc[행,열], iloc[행] 두가지임으로 열만 가져오고싶을때도 행을 비울 수 없음.
```

```python
df_concat1 = pd.read_csv("../data/concat_1.csv")
df_concat2 = pd.read_csv("../data/concat_2.csv")
df_concat3 = pd.read_csv("../data/concat_3.csv")
```

```python
df_concat1
```

```python
df_concat2
```

```python
df_concat3
```

```python
total_concat = pd.concat([df_concat1, df_concat2, df_concat3], sort=True, ignore_index=True)
```

```python
total_concat
```

```python
total_concat.iloc[0:1]  #행 인덱스0부터0까지
```

```python
total_concat.iloc[:,4]  #index=4열
```

```python
total_concat.iloc[0]  # index가 0인 행 인스턴스
```

```python
total_concat.iloc[0:5]
```

```python
total_concat.iloc[:,0:2]  #index=0.1인 열 슬라이싱
```

```python
row_append = df_concat1.append(df_concat2, ignore_index=True, sort=False)
row_append
```

```python
row_append.columns
```

```python
inner_result = pd.concat([df_concat1,df_concat2,df_concat3], join ='inner') # 공통적인 항목만 가져옴
inner_result
```

```python
inner_result2 = pd.concat([df_concat1,df_concat3], join ='inner')
inner_result2
```
