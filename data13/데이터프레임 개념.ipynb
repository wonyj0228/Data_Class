# dic사용하기

```python
dics = {"name" : ["홍길동","박길동","김길동","송길동"], "age" : [10,20,30,40]}
df = pd.DataFrame(dics)
df
```

```python
# enumerate 연습

for i, item in enumerate(df['name'],1):
     print(i," : ", item, '님')
```
