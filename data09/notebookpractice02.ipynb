```python
import numpy as np 
import pandas as pd
```

```python
not_exercise = pd.read_excel('../data09/reason_not_exercise.xls')
not_exercise
```

```python
not_exercise.drop(columns='기간', inplace=True)
```

```python
not_exercise
```

```python
not_ex_age= not_exercise[not_exercise['대분류'] == '연령별'].copy()
not_ex_age
```

```python
not_ex_age.set_index('분류',inplace=True)
not_ex_age
```

```python
not_ex_age.drop(columns='대분류',inplace=True)
```

```python
not_ex_age
```

```python
import matplotlib.pyplot as plt
```

```python
from matplotlib import font_manager, rc #한글처리
```

```python
f_path = 'C:/Windows/Fonts/HANDotum.ttf'
font_name = font_manager.FontProperties(fname=f_path).get_name()
rc('font',family = font_name)
```

```python
plt.rcParams['axes.unicode_minus'] = False
```

```python
# plot(행,열,그래프사이즈)
f, ax = plt.subplots(1, 2, figsize = (12,10))
#pit() = 파이차트를 그려라
# 첫번째 축에 들어갈 파이차트를 소수점첫째짜리까리 데이터를 적용해서 그려라
not_ex_age['운동을 할 충분한 시간이 없어서'].plot.pie(explode = [0.02, 0.02, 0.1, 0.02, 0.02, 0.02], ax = ax[0], autopct='%1.1f%%')
not_ex_age['함께 운동을 할 사람이 없어서'].plot.pie(explode = [0.02, 0.02, 0.02, 0.02, 0.02, 0.1], ax = ax[1], autopct='%1.1f%%')
ax[0].set_title('운동할 시간이 없음')
ax[1].set_title('운동할 사람이 없음')
plt.show()
```
