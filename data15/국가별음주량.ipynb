```python
cols = ['beer_servings','spirit_servings','wine_servings','total_litres_of_pure_alcohol']
```

```python
corr=drinks[cols].corr(method = 'pearson') #상관지수구하기, 피어슨상관!
corr # -1~1
```

```python
# beer_servings : 나라별 맥주소비량
# spirit_servings : 나라별 spirite소비량
# wine_servings : 나라별 와인소비량
# total_litres_of_pure_alcohol : 나라별 알콜소비량
```

* 맥주 소비량은 총 알코올 소비량과 높은 상관도가 있다. 즉 맥주소비량이 늘어날 수록 총 알코올 소비량이 높아진다.
* 세계적으로 주류 중 맥주 소비량이 가장 많다.
* 와인을 많이 마시는 사람들이 반드시 맥주를 많이 마시는 것은 아니지만 아예 관계가 없진않다.
* 스프라이트는 와인보다는 맥주와 좀 더 상관이 있다.

```python
# 대륙별 spirit_servings의 평균, 최소, 최대, 합계를 시각화합니다.
n_groups = len(result.index)
means = result['mean'].tolist()
mins = result['min'].tolist()
maxs = result['max'].tolist()
sums = result['sum'].tolist()
 
index = np.arange(n_groups) # x축 5개 생성
bar_width = 0.1
 
rects1 = plt.bar(index, means, bar_width,
                 color='r',
                 label='Mean')
 
rects2 = plt.bar(index + bar_width, mins, bar_width,
                 color='g',
                 label='Min')

rects3 = plt.bar(index + bar_width * 2, maxs, bar_width,
                 color='b',
                 label='Max')
 
rects3 = plt.bar(index + bar_width * 3, sums, bar_width,
                 color='y',
                 label='Sum')

plt.xticks(index, result.index.tolist())
plt.legend()
plt.show()
```

```python
africa = drinks.loc[drinks['continent'] == 'AF']
europe = drinks.loc[drinks['continent'] == 'EU']
```

```python
from scipy import stats
# 통계라이브러리 stats = statistics
```

```python
stats.ttest_ind(africa['beer_servings'],europe['beer_servings'])
```

```python
# africa와 europe가의 평균의 차이가 의미가 있다. 대립가설이 채택!
# 대립가설 : 아프리카, 유럽간 평균의 차이가 있다.
# 영가설 : 아프리카, 유럽간 평균의 차이가 없다. (가능성 0.05미만인 경우)
# pvalue = 가능성지수.낮을수록 대립가설이 채택됨.
```
