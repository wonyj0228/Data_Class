```python
# ë¶„ë¥˜
from sklearn.tree import DecisionTreeClassifier

# ì •í™•ë„ê°€ 100í¼ì„¼íŠ¸ê°€ ë˜ë©´ì•ˆë¨. í•™ìŠµë˜ì§€ì•Šì€ ìƒí™©ì— ëŒ€ì²˜í•´ì•¼í•˜ê¸° ë•Œë¬¸ì—.
# í•™ìŠµê¸° ê°ì²´ ìƒì„±
# íŒŒë¼ë©”í„°ë¥¼ ë¶„ì„ê°€ê°€ ìž„ì˜ë¡œ ë³€ê²½í•  ìˆ˜ ìžˆìŒ.
# íŒŒë¼ë©”í„°ë¥¼ ë¶„ì„ê°€ê°€ ë³€ê²½ê°€ëŠ¥, ë³€ê²½ ë¶ˆê°€ëŠ¥
# ë¶„ì„ê°€ê°€ ë³€ê²½ê°€ëŠ¥í•œ íŒŒë¼ë©”í„°ê°€ í•˜ì´í¼íŒŒë¼ë©”í„°
# ë°‘ì€ í•˜ì´í¼íŒŒë¼ë©”í„°!
tree_clf = DecisionTreeClassifier(max_depth = 3, random_state = 13) # íŠ¸ë¦¬ì˜ ìµœëŒ€ ê¹Šì´ = 3ë²ˆ. ëžœë¤ ì‹œë“œ.

# ê³µë¶€ì‹œí‚¤ê¸°
tree_clf.fit(X_train, Y_train)

# trainë°ì´í„°ë¥¼ ê°€ì§€ê³  ê³µë¶€í•œ ì ìˆ˜
tree_clf.score(X_train,Y_train)
```
