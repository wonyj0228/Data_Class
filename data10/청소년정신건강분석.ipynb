```python
#컬럼명을 원하는 것으로 변경하기
col_names = ['스트레스전체','스트레스남학생','스트레스여학생','우울전체','우울남학생','우울여학생','자살전체','자살남학생','자살여학생']
teen = pd.read_excel('../data10/teenage_mental.xls',
                     header = 1,
                     usecols='C:K',
                     names = col_names
                    )
teen
```

```python
# 행을 만들기
teen.loc[1] = 100 - teen.loc[0]
teen
```

```python
teen['응답'] = ['그렇다','아니다']
```

```python
teen
```

```python
teen.set_index('응답', inplace = True)
teen
```

```python
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc #한글처리
f_path = 'C:/Windows/Fonts/HANDotum.ttf'
font_name = font_manager.FontProperties(fname=f_path).get_name()
rc('font',family = font_name)
plt.rcParams['axes.unicode_minus'] = False
```

```python
teen['스트레스전체'].plot.pie(explode = [0.05,0.05])
```

```python
f, ax = plt.subplots(1,3, figsize = (15,10))
teen['스트레스전체'].plot.pie(explode=[0,0.05], ax = ax[0], autopct = '%1.1f%%')
ax[0].set_title('스트레스를 받은 적이 있다.')

teen['우울전체'].plot.pie(explode=[0,0.05], ax = ax[1], autopct = '%1.1f%%')
ax[1].set_title('우울증을 경험한 적이 있다.')

teen['자살전체'].plot.pie(explode=[0,0.05], ax = ax[2], autopct = '%1.1f%%')
ax[2].set_title('자살을 고민한 적이 있다.')

plt.show()
```

```python
f, ax = plt.subplots(1,2, figsize = (15,10))
teen['스트레스남학생'].plot.pie(explode=[0,0.02], ax = ax[0], autopct = '%1.1f%%')
ax[0].set_title('스트레스를 받은 적이 있다 남학생 응답')

teen['스트레스여학생'].plot.pie(explode=[0,0.02], ax = ax[1], autopct = '%1.1f%%')
ax[1].set_title('스트레스를 받은 적이 있다 여학생 응답')
```
